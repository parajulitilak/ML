```python
#header parameter
pd.read_csv('movielist.csv', header = 0)
```

```python
#squeeze parameters
pd.read_csv('movielist.csv', usecols = ['id'], squeeze = True)
```

```python
#skiprows example
pd.read_csv('movielist.csv', skiprows = [1,2])
```

```python
#nrows example
pd.read_csv('movielist.csv',nrows = 100)
```

```python
#encoding parameters
pd.read_csv('movielist.csv', encoding = 'latin-1')
```

```python
#dtypes parameters
pd.read_csv('movielist.csv',dtype = {'popularity':int})
```

```python
#handling dates 
pd.read_csv('movielist.csv', parse_dates = ['release_date']).info()
```

```python
#convertors
def rename(title):
    if title == "The Godfather":
        return "gf"
    else:
        return title
```

```python
pd.read_csv('movielist.csv', converters = {'title':rename})
```

```python
#na_values parameters
pd.read_csv('movielist.csv', na_values = ['title',])
```

```python
#loading a huge datasets in chunks
dfs = pd.read_csv('movielist.csv', chunksize = 500)

for chunks in dfs:
    print(chunks.size)
```
