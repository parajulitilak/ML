```python
import pandas as pd 
import seaborn as sns
```

```python
#built in dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#titanic dataset from kaggle
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
#builtin dataset
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
#popular iris builtin dataset
iris = sns.load_dataset('iris')
```

# 1.scatterplot  ( numerical - numerical )

```python
tips.head()
```

```python
df = tips
sns.scatterplot(x = tips['total_bill'],y = tips['tip'],hue=df['sex'],style=df['smoker'],size=df['size'])
```

# 2.bar plot ( Numerical - Categorical )

```python
titanic.head()
```

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Sex'], hue = titanic['Survived'])
```

# 3.Boxplot ( numerical - categorical )

```python
sns.boxplot(x =titanic['Sex'], y = titanic['Age'], hue = titanic['Pclass'])
```

# 4.Distplot ( numerical - categorical )

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'], hist = True)
sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist = True)
```

# 5.heatmap ( categorical - categorical )

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Embarked').mean()['Survived']*100)
```

# 6.clustermap ( categorical - categorical )

```python
titanic.head()
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Pclass']))
```

# 7.pairplot ( collection of scatterplot )

it automatically detects how many columns are there in your dataset and plots a scater plot between the columns.

```python
iris.head()
```

```python
sns.pairplot(iris,hue = 'petal_width')
```

# lineplot ( numerical - numerical )

special case of scatterplot
---
if we add all points of scatterplot that is called lineplot
---
if in x-axis : it is time based number that may be time,date,year,month.
---

```python
flights.head()
```

```python
new = flights.groupby('year').sum().reset_index()
sns.lineplot(x = new['year'], y = new['passengers'])
```

```python
sns.heatmap(flights.pivot_table(values = 'passengers', index ='month', columns = 'year'))
```
